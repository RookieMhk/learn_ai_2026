## pandas

```python
import pandas as pd
import numpy as np
```

```python
# 创建模拟电商用户行为数据
电商数据 = {
    '用户ID': range(1, 1001),
    '会话ID': [f'sess_{i}' for i in range(1, 1001)],
    '页面浏览量': np.random.randint(1, 20, 1000),
    '网站停留时间': np.random.randint(10, 300, 1000),
    '跳出率': np.random.uniform(0, 1, 1000),
    '是否购买': np.random.choice([0, 1], 1000, p=[0.7, 0.3]),
    '购买金额': np.where(np.random.choice([0, 1], 1000, p=[0.7, 0.3]) == 1,
                               np.random.uniform(10, 1000, 1000), 0),
    '设备类型': np.random.choice(['手机', '桌面', '平板'], 1000),
    '日期': pd.date_range('2024-01-01', periods=1000, freq='H')
}

电商数据框 = pd.DataFrame(电商数据)
print('电商用户行为数据：')
电商数据框.head()

# 练习：
# 1. 计算每个用户的平均页面浏览量
# 2. 计算不同设备的转化率（购买率）
# 3. 分析购买用户与非购买用户的行为差异
# 4. 提取消费频次、客单价等核心特征
```

```python
# 1. 计算每个用户的平均页面浏览量
电商数据框.groupby('用户ID').agg({"页面浏览量":"mean"}).reset_index()
```

```python
# 2. 计算不同设备的转化率（购买率）
电商数据框.groupby("设备类型")['是否购买'].mean()
```

```python
# 3. 分析购买用户与非购买用户的行为差异
电商数据框.groupby("是否购买").agg({"页面浏览量": np.mean, "网站停留时间": np.mean, "跳出率": np.mean}).reset_index()
```
