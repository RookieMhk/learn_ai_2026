# Matplotlib 学习笔记

## 核心目标
- 掌握 Matplotlib 基础操作
- 学习简单可视化的核心方法
- 用于 AI 开发中的结果验证

```python
import matplotlib.pyplot as plt
import numpy as np

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

## 1. 基础绘图

Matplotlib 的基本绘图功能。

```python
# 绘制折线图
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.title('正弦函数')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## 2. 多种图表类型

Matplotlib 支持多种图表类型。

```python
# 1. 散点图
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.title('散点图')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='颜色')
plt.show()

# 2. 条形图
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 34]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('条形图')
plt.xlabel('类别')
plt.ylabel('数值')
plt.show()

# 3. 直方图
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='green')
plt.title('直方图')
plt.xlabel('值')
plt.ylabel('频率')
plt.show()

# 4. 饼图
labels = ['类别1', '类别2', '类别3', '类别4']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 突出显示第二个扇形

plt.figure(figsize=(10, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('饼图')
plt.axis('equal')  # 确保饼图是圆的
plt.show()
```

## 3. 多子图

在一个 figure 中创建多个子图。

```python
# 创建 2x2 的子图
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 第一个子图：折线图
x = np.linspace(0, 2 * np.pi, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('正弦函数')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].grid(True)

# 第二个子图：散点图
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
axes[0, 1].scatter(x_scatter, y_scatter)
axes[0, 1].set_title('散点图')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')

# 第三个子图：条形图
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('条形图')
axes[1, 0].set_xlabel('类别')
axes[1, 0].set_ylabel('数值')

# 第四个子图：直方图
data = np.random.randn(1000)
axes[1, 1].hist(data, bins=20)
axes[1, 1].set_title('直方图')
axes[1, 1].set_xlabel('值')
axes[1, 1].set_ylabel('频率')

plt.tight_layout()  # 自动调整子图布局
plt.show()
```

## 4. 图表样式与自定义

自定义图表样式，使其更加美观。

```python
# 自定义折线图样式
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize=(12, 6))
plt.plot(x, np.sin(x), label='sin(x)', color='red', linewidth=2, linestyle='--', marker='o', markersize=5)
plt.plot(x, np.cos(x), label='cos(x)', color='blue', linewidth=2, linestyle='-', marker='s', markersize=5)
plt.title('正弦和余弦函数', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 2 * np.pi)
plt.ylim(-1.2, 1.2)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])
plt.show()
```

## 5. 三维图表

Matplotlib 也支持三维图表。

```python
from mpl_toolkits.mplot3d import Axes3D

# 创建三维散点图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

ax.scatter(x, y, z, c=colors, s=sizes, alpha=0.5)
ax.set_title('三维散点图')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 6. 实战应用

Matplotlib 在 AI 开发中的实际应用。

```python
# 示例1: 模型训练过程可视化
epochs = range(1, 101)
train_loss = np.random.randn(100).cumsum() + 100
train_loss = np.abs(train_loss)
val_loss = np.random.randn(100).cumsum() + 90
val_loss = np.abs(val_loss)

plt.figure(figsize=(12, 6))
plt.plot(epochs, train_loss, label='训练损失')
plt.plot(epochs, val_loss, label='验证损失')
plt.title('模型训练损失曲线')
plt.xlabel('epoch')
plt.ylabel('损失值')
plt.legend()
plt.grid(True)
plt.show()

# 示例2: 分类结果混淆矩阵可视化
confusion_matrix = np.array([[85, 5, 10], [3, 90, 7], [8, 4, 88]])
classes = ['类别1', '类别2', '类别3']

plt.figure(figsize=(10, 8))
plt.imshow(confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('混淆矩阵')
plt.colorbar()
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

# 在矩阵中显示数值
for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, confusion_matrix[i, j], ha='center', va='center', color='white' if confusion_matrix[i, j] > 50 else 'black')

plt.tight_layout()
plt.ylabel('真实标签')
plt.xlabel('预测标签')
plt.show()

# 示例3: 特征重要性可视化
features = ['特征1', '特征2', '特征3', '特征4', '特征5']
importances = [0.25, 0.30, 0.15, 0.10, 0.20]

plt.figure(figsize=(12, 6))
plt.bar(features, importances)
plt.title('特征重要性')
plt.xlabel('特征')
plt.ylabel('重要性')
plt.ylim(0, 0.35)
plt.show()
```

## 7. 图表导出

将图表导出为图片文件。

```python
# 绘制一个简单的图表
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('正弦函数')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 导出为不同格式
plt.savefig('sin_function.png', dpi=300, bbox_inches='tight')
plt.savefig('sin_function.pdf', bbox_inches='tight')
plt.savefig('sin_function.svg', bbox_inches='tight')

print('图表导出完成！')
plt.show()
```
