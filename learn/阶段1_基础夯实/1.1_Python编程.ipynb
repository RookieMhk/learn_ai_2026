## 💻 代码示例


```python
# 基础语法示例
# 变量和数据类型
name = "AI学习者"
age = 25
height = 1.75
is_student = True

# 运算符
result = 10 + 5 * 2
print(f"计算结果: {result}")

# 流程控制示例
if age >= 18:
    print("成年人")
else:
    print("未成年人")

# 循环示例
for i in range(5):
    print(f"循环次数: {i}")

# 数据结构示例
# 列表
fruits = ["苹果", "香蕉", "橙子"]
fruits.append("葡萄")
print(f"水果列表: {fruits}")

# 字典
student = {"name": "张三", "age": 20, "major": "计算机科学"}
print(f"学生信息: {student}")

# 函数示例
def greet(name):
    """问候函数"""
    return f"你好, {name}!"

print(greet("AI学习者"))

# 面向对象示例
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "汪汪!"

class Cat(Animal):
    def speak(self):
        return "喵喵!"

dog = Dog("旺财")
cat = Cat("咪咪")
print(f"狗说: {dog.speak()}")
print(f"猫说: {cat.speak()}")

# 异常处理示例
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"错误: {e}")
finally:
    print("程序执行完毕")
```

## ✅ 阶段检验


```python
# 阶段检验测试题

# 测试1: 基础语法
def test_basic_syntax():
    # 变量定义
    x = 10
    y = 20
    assert x + y == 30, "基础加法测试失败"
    
    # 条件语句
    if x > 5:
        result = "大于5"
    else:
        result = "小于等于5"
    assert result == "大于5", "条件语句测试失败"
    
    print("✅ 基础语法测试通过")

# 测试2: 数据结构
def test_data_structures():
    # 列表测试
    lst = [1, 2, 3, 4, 5]
    assert len(lst) == 5, "列表长度测试失败"
    
    # 字典测试
    dct = {"name": "测试", "value": 100}
    assert dct["name"] == "测试", "字典访问测试失败"
    
    print("✅ 数据结构测试通过")

# 测试3: 函数
def test_functions():
    def add(a, b):
        return a + b
    
    assert add(1, 2) == 3, "函数测试失败"
    print("✅ 函数测试通过")

# 运行所有测试
if __name__ == "__main__":
    test_basic_syntax()
    test_data_structures()
    test_functions()
    print("所有测试通过！Python基础掌握良好。")
```
